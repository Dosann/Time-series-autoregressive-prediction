```python
def Merge2(left_list, right_list):
    left_list_length = len(left_list)
    right_list_length = len(right_list)
    if left_list_length >= 2:
        divide = int(left_list_length/2)
        left_list = Merge2(left_list[:divide], left_list[divide:])
    if right_list_length >= 2:
        divide = int(right_list_length/2)
        right_list = Merge2(right_list[:divide], right_list[divide:])
    print('left: {}, right: {}'.format(left_list_length, right_list_length))
    if left_list_length == 1 and right_list_length == 1
        print('merged')
        return pd.merge(left = left_list[0], left_index = True, right = right_list[0], right_index = True, how = 'outer')
    
        
dataprices = Merge2(data[:int(len(data)/2)], data[int(len(data)/2):])
```
